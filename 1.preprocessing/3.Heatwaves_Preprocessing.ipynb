## 2. Heatwaves Characteristics Dataset

The cleaning steps for the Heatwaves Characteristics Dataset is as follows:
1. Visualize variables structure and NAN
2. Convert Decade into numeric.
3. Round to 3 Decimals
4. Rename variables.
5. Export dataset. 

```python
characteristics = pd.read_csv(r'C:\Users\ricar\OneDrive\Escritorio\Masters_of_Statistics\Semester_6\Modern_Data_Analytics\Project\August_Exam\MDA_August_Project\0.raw_data\Wheather_and_climate/Heat_waves/heat-waves_characteristics.csv')
characteristics = pd.DataFrame(characteristics)
```

```python
print("#==============#")
print("Data Type:")
print(characteristics.info())
print("#==============#")
print("Matrix size:")
print(characteristics.shape)
print("#==============#")
print("Missing Values:")
print(characteristics.isna().sum())
print("#==============#")
print("Dataset:")
characteristics.head()
```

```python
df_characteristics = characteristics.copy()
```

```python
Decade_Num = df_characteristics["Decade"].str.split('s',expand=True)
```

```python
df_characteristics["Decade"] = Decade_Num[0] 
df_characteristics.iloc[:, 0:1] = df_characteristics.iloc[:, 0:1].apply(pd.to_numeric, downcast='integer')
df_characteristics.iloc[:, 2:5] = round(df_characteristics.iloc[:, 2:5],3)
```

```python
df_characteristics = df_characteristics.rename(columns={"Frequency (average number of heat waves per year)":"Frequency",
                                              "Duration (average number of individual heat waves)":"Duration",
                                              "Season (average length of annual heat wave season)":"Season",
                                              "Intensity (average temperature above local threshold during heat waves)":"Intensity"})
```

```python
print(df_characteristics.info())
df_characteristics.head()
```
