```python
def bearer_oauth(r):
    """
    Method required by bearer token authentication.
    """
    r.headers["Authorization"] = f"Bearer {token}"
    r.headers["User-Agent"] = "v2FullArchiveSearchPython"
    return r
```

```python
def connect_to_endpoint(url, params):
    response = requests.request("GET", search_url, auth=bearer_oauth, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()
```

```python
# query by hashtag "heatwave", English lanquage and excluding retweets
q = "(#heatwave) lang:en -is:retweet"
```

```python
search_url = "https://api.twitter.com/2/tweets/search/all"
# query parameters:
query_params = {"query": q,
                "tweet.fields": tweet_fields, 
                "max_results": 100, # data are stored in database in porption of 100 tweets
                "start_time": start_time, 
                "end_time": end_time,
                "expansions": expansions,
               }
```

```python
def get_all_tweets(params):
    count = 0
    temp_count = 0
    next_token = ''
    while next_token is not None:
        if temp_count > 200:
            time.sleep(60*15)
            temp_count = 0
        else:
            jr = connect_to_endpoint(search_url, query_params)
            next_token = jr['meta'].get('next_token')
            query_params['next_token'] = next_token
            tweets = jr.get('data')
            ref = jr.get('includes').get('tweets')
            meta = jr.get('meta')
            errors = jr.get('errors')
            if isinstance(tweets, list):
                tar_tw.insert_many(tweets)
            if isinstance(ref,list):
                tar_ref.insert_many(ref)
            if isinstance(meta,dict):
                tar_meta.insert_one(meta)
            if isinstance(errors,list):
                tar_error.insert_many(errors)
            count+=1
            temp_count+=1
            time.sleep(30)
    return None
```

# Emotion classification for tweets

```python
# In this part, we performed emotion classification for tweets using Twitter-roBERTa-base for Emotion Recognition.
# The model is a roBERTa-base model trained on ~58M tweets and finetuned for emotion recognition (Barbieri et al, 2020).
# roBERTa-base is a transformers model pretrained on a large corpus of English data a self-supervised way.
# Source of the code can be found on Hagging Face (https://huggingface.co).
# The model predicts the scores of 4 emotions (joy, optimism, anger, sadness) for each tweet. 
# Input data is a dataset of tweets called "heatwave.json"
# The predicted emotion scores are added to input data and stored as "heatwave_v2.json"
```

```python
# import a dataset of tweets called "heatwave.json"
# here you need to change directory of heatwave.json
tweets = []
with open('/content/drive/MyDrive/Heatwaves_project/heatwave.json', 'r', encoding="utf-8") as f:
    for l in f.readlines():
        tweets.append(json.loads(l))
```

```python
# create a preprocessing function
# replace any username by '@' and any url address by http = removing non-essential information from tweet text
def preprocess(text):
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)
```

```python
# preprocessing text of tweets
for t in tweets:
    if isinstance(t.get('text'), str):
        t['cl_text'] = preprocess(t['text'])
```

### <font color='orange'> Downloading pre-trained model - Question 1</font> 

```python
# select a task: emotion classification
task='emotion'
MODEL = f"cardiffnlp/twitter-roberta-base-{task}"
# return roBERTa-base tokenizer
tokenizer = AutoTokenizer.from_pretrained(MODEL)
```

```python
# download label mapping
mapping_link = f"https://raw.githubusercontent.com/cardiffnlp/tweeteval/main/datasets/{task}/mapping.txt"
with urllib.request.urlopen(mapping_link) as f:
    html = f.read().decode('utf-8').split("\n")
    csvreader = csv.reader(html, delimiter='\t')
labels = [row[1] for row in csvreader if len(row) > 1]
```

```python
# automatically retrieve the relevant model given the name/path to the pretrained
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
# model.save_pretrained(MODEL)
```

```python
# perform emotion classification on tweets
# add additional variable "emotion" with the predicted scores
for t in tqdm(tweets):
    if 'cl_text' in t.keys():
        encoded_input = tokenizer(t['cl_text'], return_tensors='pt')
        output = model(**encoded_input)
        scores = output[0][0].detach().numpy()
        scores = softmax(scores)
        ranking = np.argsort(scores)
        ranking = ranking[::-1]
        score = {}
        for i in range(scores.shape[0]):
            l = labels[ranking[i]]
            s = scores[ranking[i]]
            score[l] = s.item()
        t['emotion'] = score
```

### <font color='orange'> NLP preprocessing and Data preparation - Question 2</font> 

```python
stopwords_en = set(stopwords.words('english'))
exclude = set(string.punctuation)

def clean_text(tweet):
  text = tweet.lower()
  text = re.sub("'", "", text) # to avoid removing contractions in english
  text = re.sub("@[A-Za-z0-9_]+","", text)
  text = re.sub("#[A-Za-z0-9_]+","", text)
  text = re.sub(r'http\S+', '', text) # remove links
  text = re.sub('[()!?]', ' ', text)
  text = re.sub('\[.*?\]',' ', text)
  text = re.sub("[^a-z0-9]"," ", text) # remove numbers
  text = re.sub(r'[0-9]+', '', text)   
  tweet = re.sub(r'\b\w{1,2}\b', '', text) # Remove words with 2 letters or less
  text = text.split()
  text = [w for w in text if not w in stopwords_en]
  text = [w for w in text if not w in exclude]
  text = " ".join(word for word in text) # remove parenthesis
  return text
```

```python
lemmatizer = WordNetLemmatizer()

# function to convert nltk tag to wordnet tag
def nltk_tag_to_wordnet_tag(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:          
        return None

def lemmatize_sentence(sentence):
    #tokenize the sentence and find the POS tag for each token
    nltk_tagged = nltk.pos_tag(nltk.word_tokenize(sentence))  
    #tuple of (token, wordnet_tag)
    wordnet_tagged = map(lambda x: (x[0], nltk_tag_to_wordnet_tag(x[1])), nltk_tagged)
    lemmatized_sentence = []
    for word, tag in wordnet_tagged:
        if tag is None:
            #if there is no available tag, append the token as is
            lemmatized_sentence.append(word)
        else:        
            #else use the tag to lemmatize the token
            lemmatized_sentence.append(lemmatizer.lemmatize(word, tag))
    return " ".join(lemmatized_sentence)
```

```python
# Example filter for noun-type structures bigrams
def bigram_filter_noun(bigram):
    tag = nltk.pos_tag(bigram)
    if tag[0][1] not in ['JJ', 'NN'] and tag[1][1] not in ['NN']:
        return False
    if bigram[0] in stopwords_en or bigram[1] in stopwords_en:
        return False
    if 'n' in bigram or 't' in bigram:
        return False
    if 'PRON' in bigram:
        return False
    return True
# Can eyeball list and choose PMI threshold where n-grams stop making sense
# In this case, get top 500 bigrams/trigrams with highest PMI score
bigram_filtered = bigram_dataframe[bigram_dataframe.apply(lambda bigram: bigram_filter_noun(bigram['bigram']) 
                                                              and bigram.pmi > 1, axis = 1)][:500]
bigrams = [' '.join(x) for x in bigram_filtered.bigram.values if len(x[0]) > 2 or len(x[1]) > 2]
```

```python
# Concatenate n-grams
def replace_ngram(x):
    for gram in bigrams:
        x = x.replace(gram, '_'.join(gram.split()))
    return x
```

### <font color='orange'> TF-IDF and Latent Dirichlet Allocation parameter tuning - Question 2</font>


```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.80, min_df=10,stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(new_sample_1['clean_text_lemmatized_3_noun'] )
df0 = pd.DataFrame(tfidf_matrix.toarray())
tfid_feature_names = tfidf_vectorizer.get_feature_names_out()
df0.columns = tfid_feature_names
df0
```

```python
params = {'n_components': [3,4,5,6], 'learning_decay': [.5,.7]}
# Init the Model
lda = LatentDirichletAllocation(random_state=0)
# Init Grid Search Class
model = GridSearchCV(lda, param_grid=params)
```

```python
%%time
model.fit(tfidf_matrix)
```

### <font color='orange'> LDA topics - Question 2</font>

```python
# Best Model
best_lda_model = model.best_estimator_
best_lda_model
```

```python
# Model Parameters
print("Best Model's Params: ", model.best_params_)
# Log Likelihood Score
print("Best Log Likelihood Score: ", model.best_score_)
# Perplexity
print("Model Perplexity: ", best_lda_model.perplexity(tfidf_matrix))
```

```python
n_top_words = 15
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharex=True)
axes = axes.flatten()

plt.subplots_adjust(wspace=0.4)

for topic_idx, topic in enumerate(model.best_estimator_.components_):
    top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
    top_features = [tfid_feature_names[i] for i in top_features_ind]
    weights = topic[top_features_ind]

    ax = axes[topic_idx]
    ax.barh(top_features, weights, height=0.7, color='darkturquoise', edgecolor='blue')
    ax.set_title(f'LDA Topic {topic_idx +1}',
                     fontdict={'fontsize': 14})
    ax.invert_yaxis()
    ax.tick_params(axis='both', which='major', labelsize=10)
```

```python
# In this part, we aimed to create a graph using co-occurrence of hashtag "heatwave" with other hashtags.
# We created an undirencted graph where the nodes represent hashtags 
# and the edges represent co-occurence between hashtags.
# Gephi was used to visualize the graph of 82 nodes with co-occurence > = 1000 times.
# https://gephi.org/
```

```python
# import a dataset of tweets
src_tw = []
with open('/content/drive/MyDrive/Heatwaves_project/heatwave.json', 'r', encoding="utf-8") as f:
    for l in f.readlines():
        src_tw.append(json.loads(l))
```

```python
# print total number of tweets
len(src_tw)
```

```python
# create an emply list for hashtags
hashtags = []
```

```python
# create a list of hashtags from metadata (['entities']['hashtags'])
for i in src_tw:
    try:
        tags = i['entities']['hashtags']
        tags = [t['tag'] for t in tags]
        hashtags.append(tags)
    except:
        print(i)
```

```python
# print out hastags for any tweet (in this case tweet2)
hashtags[2]
```

```python
# function that creates all possible pairs
def get_cooccurrence(docs):
    pairs = []
    for doc in docs:
        pairs.extend(list(combinations(doc, 2)))
    return pairs
```

```python
# function that selects only unique pairs and counts them
def count_unique(pairs):
    p_count = Counter()
    for (a,b) in pairs:
        if ((a,b) or (b,a)) in p_count:
            p_count[(a,b)]+=1
        elif ((a,b) or (b,a)) not in p_count:
            p_count[(a,b)]=1
    return p_count
```

```python
prs = get_cooccurrence(hashtags)
```

```python
prs_cnt = count_unique(prs)
```

```python
prs_w = []
for i in prs_cnt.keys():
    prs_w.append((*i,float(prs_cnt[i])))
```

```python
# create an undirected graph
G = nx.MultiGraph()
G.add_weighted_edges_from(prs_w)
```

```python
# write a graph in graph exchange XML format 
nx.write_gexf(G, "/content/drive/MyDrive/Heatwaves_project/heatwave.gexf")
```
